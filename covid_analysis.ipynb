```python
map_covid_variants = {
    "alpha":"B.1.1.7",
    "beta":"B.1.351",
    "gamma":"P.1",
    "delta":"B.1.617",
    "delta.1":"B.1.617.1",
    "delta.3":"B.1.617.3",
    "delta.2":"B.1.617.2",
    "omicron":"B.1.1.529",
    "lambda":"C.37",
    "mu":"B.1.621",
    "delta+":"AY.4.2",
    "ihu":"B.1.640",
    "ihu.2":"B.1.640.2"
}
```

```python
def replace(df, col, key, val):
    m = [v == key for v in df[col]]
    df.loc[m, col] = val
```

```python
for var, code in map_covid_variants.items():
    replace(df, "variant", code, var)
```
